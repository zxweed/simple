## cython class interaction demo

```python
import cython
%load_ext cython
```

```python
%%cython -a
#distutils: language=c++

cdef class Book:
    cdef public int bar
    
    def __init__(self):
        self.bar = 4
        print('init bar=', self.bar)
        
    cdef increment(self):
        self.bar += 1
        
    def process(self):
        print('process')
        for _ in range(100_000_000):
            self.bar += 1
            #self.increment()
```

```python
book = Book()
%time book.process()
book.bar
```

```python
book.process()
book.bar
```
